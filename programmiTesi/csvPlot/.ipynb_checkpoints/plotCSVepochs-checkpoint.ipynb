```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


for epoch in range(300,1200+150,150):
  #print(epoch)
  try:
    arr=np.genfromtxt(f"drive/MyDrive/csvsANIME/csv{epoch}.csv",delimiter=',')
  except:
    continue
  plt.scatter(arr[0,:],arr[1,:]/arr[0,:],label=f"{epoch}")
plt.legend(loc="best")
plt.subplots_adjust(right=1.5)
plt.plot(arr[0,:],[0.5 for i in range (8)],"--",color='k')
plt.xlabel("top k")
plt.ylabel("percentage of correctly identified samples")
plt.show()
```
